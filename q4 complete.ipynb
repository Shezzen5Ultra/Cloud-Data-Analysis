#### Function to load data

```python
#export
def load_data(gcp_storage_path: str) -> pyspark.sql.DataFrame:
    """
        :param gcp_storage_path: string (full gs path including file name e.g gs://bucket_name/data.csv) 
        :return: spark dataframe  
    """
    df = spark.read.format("csv").option("header", True).option("inferSchema", True).load(gcp_storage_path)
    return df 
```

#### Function to exclude trips that don't have a pickup location

```python
#export
def exclude_no_pickuplocations(df: pyspark.sql.DataFrame) -> pyspark.sql.DataFrame:
    """
        :param nyc tax trips dataframe: spark dataframe 
        :return: spark dataframe  
    """
    pickup_location_col = "pulocationid"
    filtered_df = df.filter(df[pickup_location_col].isNotNull() & (df[pickup_location_col] != 0))
    return filtered_df
```

#### Function to exclude trips with no distance

```python
#export
def exclude_no_tripdistance(df: pyspark.sql.DataFrame) -> pyspark.sql.DataFrame:
    """
        :param nyc tax trips dataframe: spark dataframe 
        :return: spark dataframe  
    """
    trip_distance_col = "trip_distance"
    filtered_df = df.filter(df[trip_distance_col].isNotNull() & (df[trip_distance_col] != 0))
    return filtered_df
```

#### Function to include fare amount between the range of 20 to 60 Dollars

```python
#export
def include_fare_range(df: pyspark.sql.DataFrame) -> pyspark.sql.DataFrame:
    
    """
        :param nyc tax trips dataframe: spark dataframe 
        :return: spark dataframe  
    """
    fare_col = "fare_amount" 
    filtered_df = df.filter((df[fare_col] >= 20) & (df[fare_col] <= 60))
    return filtered_df
```

#### Function to get the highest tip amount

```python
#export
def get_highest_tip(df: pyspark.sql.DataFrame):
    """
        :param nyc tax trips dataframe: spark dataframe 
        :return: decimal (rounded to 2 digits)  (NOTE: DON'T USE FLOAT)
    """
    tip_col = "tip_amount"
    highest_tip_row = df.select(df[tip_col]).orderBy(df[tip_col].desc()).limit(1)
    highest_tip = highest_tip_row.first()[0]
    highest_tip_rounded = round(highest_tip, 2)
    return highest_tip_rounded
```

#### Function to get total toll amount

```python
#export
def get_total_toll(df: pyspark.sql.DataFrame):
    """
        :param nyc tax trips dataframe: spark dataframe 
        :return: decimal (rounded to 2 digits)  (NOTE: DON'T USE FLOAT)
    """
    toll_col = "tolls_amount"

    toll_amounts = df.select(df[toll_col]).collect()
    if not toll_amounts:
      return 0.0
    total_toll = sum(row[0] for row in toll_amounts)
    total_toll_rounded = round(total_toll, 2)
    return total_toll_rounded
```
